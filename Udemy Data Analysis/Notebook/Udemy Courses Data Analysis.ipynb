### 7. Show all courses of Graphic Design where the price is below 100 ?

```python
df.head(3)
```

```python
filtered = df[(df.subject == 'Graphic Design') & (df.is_paid == True)]
filtered
```

```python
filtered.dtypes
```

```python
filtered = filtered.astype({'price':'int'})
```

```python
filtered.dtypes
```

```python
filtered.shape
```

```python
final_df = filtered[filtered.price < 100]
final_df
```

```python
final_df.shape
```

```python
free_graphic = df[(df.subject == 'Graphic Design') & (df.is_paid == False)]
free_graphic
```

```python
free_graphic.shape
```

```python
pd.concat([free_graphic, final_df], ignore_index=True)
```

### 10. What are the Max. Number of Subscribers for Each Level of courses ?

```python
df.groupby('level')['num_subscribers'].max().reset_index().sort_values('num_subscribers', ascending=False)
```
